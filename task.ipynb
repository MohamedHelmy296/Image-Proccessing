```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
plant = cv2.imread("plant.png")
plt.imshow(cv2.cvtColor(plant,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
hist = cv2.calcHist([plant],[0], None, [256], [0,256])
```

```python
intensity_values = np.array([x for x in range(hist.shape[0])])
```

```python
color = ('blue','green','red')
for i,col in enumerate(color):
    histr = cv2.calcHist([plant],[i],None,[256],[0,256])
    plt.plot(intensity_values,histr,color = col,label=col+" channel")
    
    plt.xlim([0,256])
plt.legend()
plt.title("Histogram Channels")
plt.show()
```

```python
def plot_image(image_1, image_2,title_1="Orignal", title_2="New Image"):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
```

```python
alpha = 1 # Simple contrast control
beta = -200  # Simple brightness control   
new_image = cv2.convertScaleAbs(plant, alpha=alpha, beta=beta)
```

```python
plt.imshow(cv2.cvtColor(new_image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(cv2.cvtColor(new_image,cv2.COLOR_BGR2RGB))
plt.show()
plot_image(plant, new_image, "Orignal", "brightness & contrast control")
```

```python
color = ('blue','green','red')
for i,col in enumerate(color):
    histr = cv2.calcHist([new_image],[i],None,[256],[0,256])
    plt.plot(intensity_values,histr,color = col,label=col+" channel")
    
    plt.xlim([0,256])
plt.legend()
plt.title("Histogram Channels")
plt.show()
```
